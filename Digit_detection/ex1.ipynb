```python
import tensorflow as tf
import tensorflow.examples.tutorials.mnist.input_data as input_data
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2

# Create model
def multilayer_perceptron(net_input, weights, biases):
    # First layer with sigmoid activation
    layer_1 = tf.add(tf.matmul(net_input, weights['h1']), biases['b1'])
    layer_1 = tf.nn.sigmoid(layer_1)
    # Hidden layer with sigmoid activation
    layer_2 = tf.add(tf.matmul(layer_1, weights['h2']), biases['b2'])
    layer_2 = tf.nn.sigmoid(layer_2)
    # Hidden layer with sigmoid activation
    layer_3 = tf.add(tf.matmul(layer_2, weights['h3']), biases['b3'])
    layer_3 = tf.nn.sigmoid(layer_3)
    # Output layer with softmax activation
    out_layer = tf.matmul(layer_3, weights['out']) + biases['out']
    out_layer = tf.nn.softmax(out_layer)
    return out_layer

# read the data and labels as ont-hot vectors
# one-hot means a sparse vector for every observation where only
# the class label is 1, and every other class is 0.
mnist = input_data.read_data_sets('MNIST_data/', one_hot=True)

# Parameters
batch_size = 300
n_epochs = 250
eta = 0.01375
model_path = "saved_model/model.ckpt"

# Network Parameters
n_input = 784 # MNIST data input (img shape: 28*28)
n_hidden_1 = 256 # 1st layer number of features
n_hidden_2 = 256 # 2nd layer number of features
n_hidden_3 = 64 # 3rd layer number of features
n_classes = 10 # MNIST total classes (0-9 digits)

net_input = tf.placeholder(tf.float32, [None, n_input], name="input")
y_true = tf.placeholder(tf.float32, [None, n_classes])

# Store layers weight & bias
weights = {
    'h1': tf.Variable(tf.random_normal([n_input, n_hidden_1])),
    'h2': tf.Variable(tf.random_normal([n_hidden_1, n_hidden_2])),
    'h3': tf.Variable(tf.random_normal([n_hidden_2, n_hidden_3])),
    'out': tf.Variable(tf.random_normal([n_hidden_3, n_classes]))
}
biases = {
    'b1': tf.Variable(tf.random_normal([n_hidden_1])),
    'b2': tf.Variable(tf.random_normal([n_hidden_2])),
    'b3': tf.Variable(tf.random_normal([n_hidden_3])),
    'out': tf.Variable(tf.random_normal([n_classes]))
}

# Construct model
net_output = multilayer_perceptron(net_input, weights, biases)

# Gives the net output name for the save model and load
net_output = tf.identity(net_output, name="output")

# prediction and actual using the argmax as the predicted label
correct_prediction = tf.equal(tf.argmax(net_output, 1), tf.argmax(y_true, 1))

# And now we can look at the mean of our network's correct guesses
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))

# Sets the cost funcation 
cost = tf.reduce_sum(tf.nn.softmax_cross_entropy_with_logits(logits=net_output, labels=y_true))

# Using adam optimizer
optimizer = tf.train.AdamOptimizer(eta).minimize(cost)

# Creating a session
sess = tf.Session()
sess.run(tf.global_variables_initializer())

# Inits list for the graphes
l_accuracies_train = list()
l_accuracies_validation = list()
l_cost_train = list()
l_cost_valid = list()

# Disabling limition on the amount of models to save
saver = tf.train.Saver(max_to_keep=None)

# Sets parameter that we want to save the model every <X> epochs
save_gap_between_epochs = 50

# Training the model
for epoch_i in range(n_epochs):
    for batch_i in range(0, mnist.train.num_examples, batch_size):
        batch_xs, batch_ys = mnist.train.next_batch(batch_size)
        sess.run(optimizer, feed_dict={
            net_input: batch_xs,
            y_true: batch_ys
        })      
        
    # If we need to save the model (by our own parameters)
    if (epoch_i + 1) % save_gap_between_epochs == 0:
    # Saves the model so we can load it later    
    saver.save(sess, model_path + str(epoch_i + 1))    
        
    #-----------------------Sets cost and accuracy data for graphes-------------------    
        
    # Gets the loss data of the train set
    cost_func_train = sess.run(cost, feed_dict={
                       net_input: mnist.train.images,
                       y_true: mnist.train.labels })
    
    # Normalize the cost
    cost_normalized_train = cost_func_train / mnist.train.num_examples
    
    print('Validation cost for train epoch {} is: {}'.format(epoch_i + 1, cost_normalized_train))
    
    # Adds the normalized cost to the train cost list
    l_cost_train.append(cost_normalized_train)
    
    # Gets the loss data of the validation set
    cost_func_valid = sess.run(cost, feed_dict={
                       net_input: mnist.validation.images,
                       y_true: mnist.validation.labels })
    
    # Normalize the cost
    cost_normalized_valid = cost_func_valid / mnist.validation.num_examples
    
    print('Validation cost for valid epoch {} is: {}'.format(epoch_i + 1, cost_normalized_valid))
    
    # Adds the normalized cost to the validation cost list
    l_cost_valid.append(cost_normalized_valid)
        
    # Gets the Accuracy of the train set            
    accuracy_func_train = sess.run(accuracy, feed_dict={
                       net_input: mnist.train.images,
                       y_true: mnist.train.labels })
    print('Validation accuracy for train epoch {} is: {}'.format(epoch_i + 1, accuracy_func_train))
    
    # Adds the accuracy to the train accuracy list
    l_accuracies_train.append(accuracy_func_train)
    
    # Gets the Accuracy of the validation set 
    accuracy_func_validation = sess.run(accuracy, feed_dict={
                       net_input: mnist.validation.images,
                       y_true: mnist.validation.labels })
    print('Validation accuracy for validation epoch {} is: {}'.format(epoch_i + 1, accuracy_func_validation))
    
    # Adds the accuracy to the validtion accuracy list
    l_accuracies_validation.append(accuracy_func_validation)
    
   
# Prints the accuracy grpah
plt.title('Logistic Regression Acuracy - Validation (Green) Vs Train (Purple)')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.plot(l_accuracies_train, color='m')
plt.plot(l_accuracies_validation, color='g')
plt.show()

# Prints the loss grpah
plt.title('Logistic Regression cost - Validation (Green) Vs Train (Purple)')
plt.ylabel('cost')
plt.xlabel('Epochs')
plt.plot(l_cost_train, color='m')
plt.plot(l_cost_valid, color='g')
plt.show()

print("Accuracy for test set: {}". format(sess.run(accuracy,
               feed_dict={
                   net_input: mnist.test.images,
                   y_true: mnist.test.labels
               })))
```
